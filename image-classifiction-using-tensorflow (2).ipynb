```python
# Create a data generator
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Load training images (since they're already resized)
train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=(224, 224),  # Keep it here for safety, but it's already 224x224
    batch_size=32,
    class_mode='binary'
)

# Load test images (since they're already resized)
validation_generator = test_datagen.flow_from_directory(
    valid_path,
    target_size=(224, 224),  # Keep it here for safety, but it's already 224x224
    batch_size=32,
    class_mode='binary'
)
```

```python
model = Sequential()

# IF you want to know the summary of you model i-e the inputs and outputs of all the layers in your
# architecture, you have to add the "Input()" layer. This specifies to tensorflow what shape of
# data should it expect
model.add(Input(shape = [224,224,3]))

# Add the pre-trained VGG16 model as the base
vgg16_base = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Add it to the Sequential model
model.add(vgg16_base)

# Check the model summary so far
model.summary()
```

```python
# Create a new Sequential model
model = Sequential()

# IF you want to know the summary of you model i-e the inputs and outputs of all the layers in your
# architecture, you have to add the "Input()" layer. This specifies to tensorflow what shape of
# data should it expect
model.add(Input(shape=[224,224,3]))

# Add the pre-trained VGG16 model as the base (already done)
model.add(vgg16_base)

# Flatten the output from VGG16
model.add(Flatten())

# Add a fully connected layer with 256 neurons and ReLU activation
model.add(Dense(256, activation='relu'))

# Add the final output layer with 1 neuron and sigmoid activation for binary classification
model.add(Dense(1, activation='sigmoid'))

# Display the complete model summary
model.summary()

```

```python
# This is another way of creating the architecture of your model (using the Sequential function)


model = tf.keras.Sequential([
    Input(shape=[224,224,3]),
    vgg16_base,
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=256,activation='relu'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])
model.summary()
```

```python
# You had set the loss to "categorgical_crossentropy", whereas the problem was binary classification
# I corrected it to "binary_crossentropy"

model.compile(
    loss = "binary_crossentropy",
    optimizer = "adam",
    metrics = ['accuracy']
)
```

```python
## data transformation

train_dataGen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True

)

test_dataGen = ImageDataGenerator(rescale = 1./255)
```

```python
# loading train images
train_set = train_dataGen.flow_from_directory(
    "/content/drive/MyDrive/computer-vision-work/cat-dogs-dataset/training_set",
    target_size = (224, 224),
    batch_size = 32,
    class_mode = 'binary'
    )

# class_mode here was categorical. I set it to binary, because we are expecting 2 outcomes
# Either cat, or dog
```

```python
# loading testing images
test_set = test_dataGen.flow_from_directory(
    "/content/drive/MyDrive/computer-vision-work/cat-dogs-dataset/test_set",
    target_size = (224, 224),
    batch_size = 32,
    class_mode = 'binary'
    )

# class_mode here was categorical. I set it to binary, because we are expecting 2 outcomes
# Either cat, or dog
```

```python
history = model.fit(
    train_set,
    validation_data = test_set,
    epochs = 25,
    steps_per_epoch = len(train_set),
    validation_steps = len(test_set)
)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['loss'], label = 'train loss')
plt.plot(history.history['val_loss'], label = 'val loss')
plt.legend()
plt.show()
plt.savefig('Loss graph')
```

```python
# now see the accuracy graph
plt.plot(history.history['accuracy'], label = 'train accuracy')
plt.plot(history.history['val_accuracy'], label = 'val accuracy')
plt.legend()
plt.show()
plt.savefig('accuracy graph')
```
